```python
def content(tag):
    if tag != None and tag.text != None and str(tag.text).strip() != "":
        return str(tag.text).strip() + " "
    else:
        return None
    
def contentText(tag):
    if content(tag) == None:
        return ""
    else:
        return content(tag)
    
def contentLabel(tag, label):
    text = contentText(tag)
    if text == "":
        return ""
    else:
        return "{}: {}, ".format(label, text)
```

```python

def xmlToDictTestHOLDING(root):

        data = {
            "submitURL": None,
            "year": None,
            "quarter": None,
            "termEffectiveDate": None,
            "orgName": None,                      
            "clientName": None,       
            "senateID": None,
            "houseID": None,
            "clientIsStateOrLocalGov": None,
            
            "name": None,
            "addr": None,
            "city": None,
            "state": None,
            "zip": None,
            "country": None,
            "diffAddress": None,
            
            "prinCity": None,
            "prinState": None,
            "prinZip": None,
            "prinZipExt": None,
            "prinCountry": None,           
            
            "SelfSelect": None,
            "SignedName": None,
            "SignedDate": None,
            "income": None,
            "noLobbying": None,
            "expenses": None,
            "expensesMethod": None, 
            
            "alis": [],
            
            "updatedClientAddr": None,
            "updatedClientCity": None,
            "updatedClientState": None,
            "updatedClientZip": None,
            "updatedClientZipExt": None,
            "updatedClientCountry": None,
            "updatedPrinClientCity": None,
            "updatedPrinClientState": None,
            "updatedPrinClientZip": None,
            "updatedPrinClientZipExt": None,          
            "updatedPrinClientCountry": None,
            "updatedClientDescription": None,   
            "foreignEntities": None,
            "inactive_lobbyists": None,         
            "affiliatedOrgs": None,
            "inactive_orgs": None,
            "affiliatedOrgUrl": None,
            "inactive_ali_codes": None,
            "inactive_foreign_ents": None         
        }

        data["submitURL"] = content(root.find('./submitURL'))    
        data["senateID"] = content(root.find('./senateID'))
        data["houseID"] = content(root.find('./houseID'))
        data["income"] = content(root.find('./income'))        
        data["noLobbying"] = content(root.find('./noLobbying'))       
        
        data["expenses"] = content(root.find('./expenses'))   
        data["expensesMethod"] = content(root.find('./expensesMethod'))
        data["printedName"] = content(root.find('./printedName'))
        data["clientName"] = content(root.find('./clientName'))
        data["clientAddr"] = content(root.find('./address1'))        
        data["orgName"] = content(root.find('./organizationName'))
        data["year"] = content(root.find('./reportYear'))
        data["quarter"] = content(root.find('./reportType'))
        
        data["termEffectiveDate"] = content(root.find('./terminationDate'))
        data["affiliatedOrgUrl"] = content(root.find('./updates/affiliatedUrl'))
        data["clientIsStateOrLocalGov"] = content(root.find('./clientGovtEntity'))
        
        data["city"] = content(root.find('./city'))
        data["state"] = content(root.find('./state'))     
        data["country"] = content(root.find('./country'))
        data["diffAddress"] = content(root.find('./registrantDifferentAddress'))
        
        data["prinCity"] = content(root.find('./principal_city'))
        data["prinState"] = content(root.find('./principal_state'))
        data["prinZip"] = content(root.find('./principal_zip'))
        data["prinZipExt"] = content(root.find('./principal_zipext'))
        data["prinCountry"] = content(root.find('./principal_country'))
        
       
        data["contactIntlPhone"] = content(root.find('./contactIntlPhone'))
        data["SelfSelect"] = content(root.find('./selfSelect'))
        data["SignedName"] = content(root.find('./printedName'))
        data["SignedDate"] = content(root.find('./signedDate'))
        
        data["updatedClientAddr"] = content(root.find('./updates/clientAddress'))
        data["updatedClientCity"] = content(root.find('./updates/clientCity'))
        data["updatedClientState"] = content(root.find('./updates/clientState'))
        data["updatedClientZip"] = content(root.find('./updates/clientZip'))
        data["updatedClientZipExt"] = content(root.find('./updates/clientZipext'))
        data["updatedClientCountry"] = content(root.find('./updates/clientCountry'))      
        data["updatedPrinClientCity"] = content(root.find('./updates/prinClientCity'))
        data["updatedPrinClientState"] = content(root.find('./updates/prinClientState'))
        data["updatedPrinClientZip"] = content(root.find('./updates/prinClientZip'))
        data["updatedPrinClientZipExt"] = content(root.find('./updates/prinClientZipext'))
        data["updatedPrinClientCountry"] = content(root.find('./updates/prinClientCountry'))
        data["updatedClientDescription"] = content(root.find('./updates/generalDescription'))        
        
        prefix = contentText(root.find('./prefix'))
        fNameText = contentText(root.find('./firstName'))
        lNameText = contentText(root.find('./lastName'))
        if fNameText != "" and lNameText != "":
            nameText = fNameText + " " + lNameText
            if prefix != "":
              nameText = prefix + " " +  nameText 
            data["name"] = nameText
            
        addr1 = contentText(root.find('./address1'))
        addr2 = contentText(root.find('./address2'))
        if addr1 != "":
            data['addr'] = addr1 + (", {}".format(addr2) if addr2 != "" else "")
            
        zip1 = contentText(root.find('./zip'))
        zip2 = contentText(root.find('./zipext'))
        if zip1 != "":
            data["zip"] = zip1 + ("-{}".format(zip2) if zip2 != "" else "")
            
        
        
                                  
                                  
                                  
                            
        
        """              
        # client_name = root.find('./clientName')        
        #if client_name != None and client_name.text != None: data["clientName"] = str(client_name.text)
        
        #client_addr = root.find('./address1')
        #if client_addr != None: data["clientAddr"] = client_addr.text
            
        #senate_id = root.find('./senateID')
        #if senate_id != None: data["senateID"] = senate_id.text
        #house_id = root.find('./houseID')
        #if house_id != None: data["houseID"] = house_id.text
        
        _income = root.find('./income')
        if _income != None: data["income"] = _income.text
        _expenses = root.find('./expenses')
        if _expenses != None: data["expenses"] = _expenses.text
        printed_name = root.find('./printedName')
        if printed_name != None: data["printedName"] = printed_name.text
        
        data["expenses"] = content(root.find('./expenses'))
        data["expenses"] = content(root.find('./expenses'))
        data["expenses"] = content(root.find('./expenses'))
        data["expenses"] = content(root.find('./expenses'))  
            
        """

                     
        inactiveLobbyists = root.findall('./updates/inactive_lobbyists/inactive_lobbyist')
        inactiveNames = ""
        for inactiveLobb in inactiveLobbyists:
            fName = contentText(inactiveLobb.find('firstName'))
            lName = contentText(inactiveLobb.find('lastName'))
            suff = contentText(inactiveLobb.find('suffix'))
            if fName != None and lName != None:
                if inactiveNames != "":
                    inactiveNames += "; "
                inactiveNames = inactiveNames + "Name: {} {}".format(fName, lName)
                if suff != "":
                    inactiveNames += (", " + suff)
        if inactiveNames != "":     
            data['inactive_lobbyists'] = inactiveNames
            
        inactiveOrgs = root.findall('./updates/inactiveOrgs/inactiveOrgName')
        inactiveOrgNames = ""
        for inactiveOrg in inactiveOrgs:
            nextOne = contentText(inactiveOrg)
            if inactiveOrgNames != "" and nextOne != "":
                inactiveOrgNames += "; "
            inactiveOrgNames += nextOne
        if inactiveOrgNames != "":
            data['inactive_orgs'] = inactiveOrgNames
                   
        affiliatedOrgs = root.findall('./updates/affiliatedOrgs/affiliatedOrg')
        affOrgText = ""
        for affiliatedOrg in affiliatedOrgs:   
            affOrgName = contentText(affiliatedOrg.find('affiliatedOrgName'))
            if affOrgName != "":
                if affOrgText != "":
                    affOrgText += "; "
                affOrgText = affOrgText + "Org Name: {}, ".format(affOrgName)
                affOrgText += contentText(affiliatedOrg.find('affiliatedOrgAddress'))
                affOrgText += contentText(affiliatedOrg.find('affiliatedOrgCity'))
                affOrgText += contentText(affiliatedOrg.find('affiliatedOrgState'))        
                affOrgText += contentText(affiliatedOrg.find('affiliatedOrgZip'))
                affOrgText += contentText(affiliatedOrg.find('affiliatedOrgCountry'))
                
                affPrinCity = contentText(affiliatedOrg.find('affiliatedPrinOrgCity'))
                affPrinState = contentText(affiliatedOrg.find('affiliatedPrinOrgState'))
                affPrinCountry = contentText(affiliatedOrg.find('affiliatedPrinOrgCountry'))
                
                if affPrinCity != "":
                    affOrgText += ("Principle Org city: {}".format(affPrinCity))
                if affPrinState != "":
                    affOrgText += ("Principle Org state: {}".format(affPrinState))
                if affPrinCountry != "":
                    affOrgText += ("Principle Org country: {}".format(affPrinCountry))                           
        if affOrgText != "":
            data["affiliatedOrgs"] = affOrgText
        
        
        foriegnEnts = root.findall('./updates/foreignEntities/foreignEntity')
        foriegnEntText = ""
        for foriegnEnt in foriegnEnts:                
            entName = contentText(foriegnEnt.find('name'))
            if entName != "":
                if foriegnEntText != "":
                    foriegnEntText += "; "
                foriegnEntText += ("Foriegn Ent Name: {}".format(entName))
                
                addr = contentText(foriegnEnt.find('address'))
                city = contentText(foriegnEnt.find('city'))
                state = contentText(foriegnEnt.find('state'))
                country = contentText(foriegnEnt.find('country'))    
                if addr != "":
                     foriegnEntText += ", {}".format(addr)
                if city != "":
                     foriegnEntText += ", {}".format(city)
                if state != "":
                     foriegnEntText += ", {}".format(state)
                if country != "":
                     foriegnEntText += ", {}".format(country)
     
                prinCity = contentText(foriegnEnt.find('prinCity'))
                prinState = contentText(foriegnEnt.find('prinState'))
                prinCountry = contentText(foriegnEnt.find('prinCountry'))
                if prinCity != "":
                    foriegnEntText += (", Principle city: {}".format(prinCity))
                if prinState != "":
                    foriegnEntText += (", Principle state: {}".format(prinState))
                if prinCountry != "":
                    foriegnEntText += (", Principle country: {}".format(prinCountry))
                                    
                contrib = contentText(foriegnEnt.find('contribution'))
                percentage = contentText(foriegnEnt.find('ownership_Percentage'))
                if contrib != "":
                   foriegnEntText += (", Contribution: {}".format(contrib)) 
                if percentage != "":
                   foriegnEntText += (", Ownership percentage: {}".format(percentage)) 
        if foriegnEntText != "":
            data["foreignEntities"] = foriegnEntText
               
        inactive_alis = root.findall('./updates/inactive_ALIs/ali_Code')
        inactive_ali_codes_text = ""
        for inactive_ali in inactive_alis:
            nextone = contentText(inactive_ali)
            if inactive_ali_codes_text != "" and nextone != "":
                inactive_ali_codes_text += ", "
            inactive_ali_codes_text += nextone
        if inactive_ali_codes_text != "":
            data['inactive_ali_codes'] = inactive_ali_codes_text
                  
        inactiveForeignEnt = root.find('./updates/inactive_ForeignEntities')
        inactive_foreignent_text = ""
        for inactive_foreignent in inactiveForeignEnt:
            nextone = contentText(inactive_foreignent)
            if inactive_foreignent_text != "" and nextone != "":
                inactive_foreignent_text += ", "
            inactive_foreignent_text += nextone
        if inactive_foreignent_text != "":
            data['inactive_foreign_ents'] = inactive_foreignent_text
                       
        alis = root.findall('./alis/ali_info')
        for ali in alis:
            ali_data = {
                "area_code": None,
                "issues": None,
                "agencies": None,
                "lobbyists": None,
                "foreign_entity_issues": None               
            }
            
            if ali.find('issueAreaCode') != None and ali.find('issueAreaCode').text != None:
               ali_data["area_code"] = str(ali.find('issueAreaCode').text)
            else:
                continue
                
            if ali.find('specific_issues') != None:
                descrs = ali.findall('./specific_issues/description')
                issue_text = ""
                for desc in descrs:
                    if desc != None and desc.text != None:
                        issue_text = "ISSUE: "
                        for line in str(desc.text).splitlines():
                            issue_text = issue_text + line + " " 
                        ali_data['issues'] = issue_text
            else:
                continue
                
            if ali.find('federal_agencies') != None and ali.find('federal_agencies').text != None:
                ali_data['agencies'] = str(ali.find('federal_agencies').text)
                
            if ali.find('lobbyists') != None and ali.find('./lobbyists/lobbyist') != None:
                lobbyistNames = ""
                lobbyists = ali.findall('./lobbyists/lobbyist')
                for lobbyist in lobbyists:

                    lobFirstName = contentText(lobbyist.find('lobbyistFirstName'))
                    lobLastName = contentText(lobbyist.find('lobbyistLastName'))
                    lobSuffix = contentText(lobbyist.find('lobbyistSuffix'))
                    lobCoveredPos = contentText(lobbyist.find('coveredPosition'))
                    
                    if lobFirstName != "" and lobLastName != "":
                        if lobbyistNames != "":
                            lobbyistNames += "; "
                        lobbyistNames += "Lobbyist name: {} {}".format(lobFirstName, lobLastName)
                        if lobSuffix != "":
                           lobbyistNames += (" " + lobSuffix)
                        if lobCoveredPos != "":
                           lobbyistNames += (", Covered Position: " + lobCoveredPos) 
                if lobbyistNames != "":
                    ali_data['lobbyists'] = lobbyistNames
            
            forEntIss = contentText(ali.find('foreign_entity_issues'))
            if forEntIss != "":
                ali_data['foreign_entity_issues'] = forEntIss
                
            data["alis"].append(ali_data)

        return data
                
            
            
```
